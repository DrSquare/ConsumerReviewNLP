```python
import os
import numpy as np
import pandas as pd
import ktrain
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
product_data_new = pd.read_csv('/content/drive/MyDrive/Project/Amazon_Review_Data/product_data_new.csv')
```

```python
product_data_new["Product_Review"]=product_data_new["Product_Review"].apply(str)
```

```python
reviews = product_data_new.Product_Review[0:]
```

## **Topic Modelling - each review by itself**

```python
%%time
tm = ktrain.text.get_topic_model(reviews, n_features=10000)
```

```python
tm.print_topics()
```

```python
%%time
tm.build(reviews) #, threshold=0.25)
```

```python
tm.print_topics(show_counts=True)
```

```python
tm.visualize_documents(doc_topics=tm.get_doctopics())
```

```python
doc_topics = tm.get_doctopics(topic_ids=[66, 73, 3])
tm.visualize_documents(doc_topics=doc_topics)
```

```python
feature = "battery life"
```

```python
tm.predict([feature])
```

```python
tm.topics[np.argmax(tm.predict([feature]))]
```

## **Topic Modelling - Reviews for each product combined at rating level**

```python
# Data preprocessing
prod_rating_comb = product_data_new.groupby(['Product_Id','Rating'])['Product_Review'].apply(lambda x: '|||'.join(x)).reset_index()
prod_rating_comb.Product_Review[0]
```

```python
reviews_comb = prod_rating_comb.Product_Review[0:]
```

```python
%%time
tm_comb = ktrain.text.get_topic_model(reviews_comb, n_features=10000)
```

```python
tm_comb.print_topics()
```

```python
%%time
tm_comb.build(reviews_comb)
```

```python
tm_comb.print_topics(show_counts=True)
```

```python
tm_comb.visualize_documents(doc_topics=tm_comb.get_doctopics())
```

```python
doc_topics_comb = tm_comb.get_doctopics(topic_ids=[6, 20, 4])
tm_comb.visualize_documents(doc_topics=doc_topics_comb)
```

```python
feature = "battery life"
```

```python
tm_comb.predict([feature])
```

```python
tm_comb.topics[np.argmax(tm_comb.predict([feature]))]
```
